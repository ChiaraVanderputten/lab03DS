```python
import mlxtend
import pandas as pd

! wget "https://github.com/dbdmg/data-science-lab/raw/master/datasets/online_retail.csv" -O online_retail.csv
import csv


db = [ ] 
l=[]
with open("online_retail.csv") as f:
 for cols in csv.reader(f): 
   if len(cols)==8 and cols[0][0]!= 'C':
     db.append(cols)

database = db[1:]
#print(database[0][1])

mappa={}
for i in database:
  if i[0] not in mappa:
    mappa[i[0]] = set() #non rischio di aggiungere più volte la stessa descrizione
  mappa[i[0]].add(i[2].strip())  

print(mappa['536365'])

tutti_Prodotti= set()

for i in mappa.values():
    tutti_Prodotti.update(i)
     
#print(len(tutti_Prodotti))

tutti_Prodotti= sorted(list(tutti_Prodotti))
print(tutti_Prodotti[0:20])

def get_matrice (mappa, tutti_Prodotti):
  item_pos={k : v for v,k in enumerate(tutti_Prodotti)}
  matrice=[]
  for x in mappa.values():
    riga = [0]* len(tutti_Prodotti)
    for i in x:
      riga[item_pos[i]]=1
    matrice.append(riga)  
  return matrice

matrice = get_matrice(mappa, tutti_Prodotti)

print(matrice[0])

import collections, numpy
a = numpy.array(matrice[0])
collections.Counter(a)

df = pd.DataFrame(data=matrice, columns=tutti_Prodotti)

```
